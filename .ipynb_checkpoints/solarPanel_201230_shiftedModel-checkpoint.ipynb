```python
# load_dtypes()
# DF 불러올 때부터 축소해서 가져오는 함수.

def load_dtypes(file_path):
    print(file_path)
    tmp = pd.read_csv(file_path, nrows=0) # 애초에 파일을 불러올 때
    col_dtypes= {}
    for col in tmp.columns:
        df = pd.read_csv(file_path, usecols=[col]) # 애초에 파일을 불러올 때
        
        dtype = df[col].dtype
        
        if (dtype == 'int8' or dtype == 'int16' or dtype == 'int32' or dtype == 'int64'):
            c_min = df[col].min()
            c_max = df[col].max()
            dtype = 'int'
            
        elif dtype == 'float32' or dtype == 'float64':
            c_min = df[col].min()
            c_max = df[col].max()
            dtype = 'float'
            
        elif dtype == 'object':
            n_unique = df[col].nunique()
            threshold = n_unique / df.shape[0]
            
        if dtype == 'int':
            if c_min > np.iinfo(np.int8).min and c_max < np.iinfo(np.int8).max:
                col_dtype = 'int8'
            elif c_min > np.iinfo(np.uint8).min and c_max < np.iinfo(np.uint8).max:
                col_dtype = 'uint8'
            elif c_min > np.iinfo(np.int16).min and c_max < np.iinfo(np.int16).max:
                col_dtype = 'int16'
            elif c_min > np.iinfo(np.uint16).min and c_max < np.iinfo(np.uint16).max:
                col_dtype = 'uint16'
            elif c_min > np.iinfo(np.int32).min and c_max < np.iinfo(np.int32).max:
                col_dtype = 'int32'
            elif c_min > np.iinfo(np.uint32).min and c_max < np.iinfo(np.uint32).max:
                col_dtype = 'uint32'
            elif c_min > np.iinfo(np.int64).min and c_max < np.iinfo(np.int64).max:
                col_dtype = 'int64'
            elif c_min > np.iinfo(np.uint64).min and c_max < np.iinfo(np.uint64).max:
                col_dtype = 'uint64'
                
        elif dtype == 'float':
            if c_min > np.finfo(np.float32).min and c_max < np.finfo(np.float32).max:
                col_dtype = 'float32'
            else:
                col_dtype = 'float64'
                
        elif dtype == 'object':
            if threshold > 0.7:
                col_dtype = 'object'
            else:
                col_dtype = 'category'
                
        col_dtypes[col] = col_dtype
        

    df = pd.read_csv(file_path, dtype=col_dtypes)
    
    return df
```

# shifted multi regression

본 아키텍쳐의 아이디어는 테스트 데이터의 7일 간의 데이터를 모두 활용하기 위해,<br>
하루씩 shift 된 값을 변수로 넣어 예측 모델에 학습시키는 것입니다.

시간변수인 hour는 그대로 두고, DHI, DNI, WS, RH, T, TARGET 을<br>
각각 48(하루), 96(이틀) ~ 288(엿새) 까지 역시프트를 시킵니다.<br>
그러면 각 행은 t일의 변수들, t-1일의 변수들 ~ t-6일의 변수들까지 가지게 됩니다.<br>

마찬가지로 테스트 데이터도 7일치의 데이터를 reshape 해서<br>
336 x 7 을 48 x 43 의 데이터 테이블로 변환하여 예측하게 합니다.<br>

```python
shifted_columns = ['DHI', 'DNI', 'WS', 'RH', 'T', 'TARGET']
lags = [48, 96, 144, 192, 240, 288]
```

```python
def shift_columns(data, shifted_columns, lags):
    
    temp = data.copy()
    for lag in lags:
        for col in shifted_columns:
            temp[col + '_lag_%s'%lag] = temp[col].shift(-lag)
            
    return temp
```

# weighted shifted multi regression : lgbm 에서는 의미 없음

본 아키텍쳐는 위에서 구성했던 shifted multi regression의 개선된 모형입니다.

아이디어는 간단합니다. SMR 모델이 7일치의 shift된 데이터를 변수로 가졌다면,<br>
당연히 t일의 데이터가 t-6일의 데이터보다 t+1일을 예측하는데 더 중요한 변수가 아닐까 하는 겁니다.

따라서 t-n 일의 데이터들에게 가중치를 곱해줘서 변수의 중요도를 낮추고자 합니다.<br>
t-(n*k) 가 될것입니다.

```python
weights = [0.9, 0.8, 0.7, 0.6, 0.5, 0.4]
```

```python
def weight_shift_columns(data, shifted_columns, lags, weights):
    
    temp = data.copy()
    for lag, weight in zip(lags, weights):
        for col in shifted_columns:
            temp[col + '_lag_%s'%lag] = temp[col].shift(-lag)*weight
            
    return temp
```

## 1. 파생변수 생성 해봅시다.

### 1.1 Hour + Minute

Hour가 시간 단위, Minute 은 30분 단위로 나뉘어져 있습니다. 의미 없는 Minute을 0.5로 치환하여 Hour에 더해줍니다.

```python
def sum_hour_minute(train):
    # 0과 30으로 되어 있는 minute을 0과 0.5로 바꿉니다.
    train.Minute = round(train.Minute * 0.017, 1)
    # hour에 minute을 더해줍니다.
    train.Hour = train.Hour + train.Minute
    
    return train
```

### 1.2 GHI(DHI + DNI*COS(k))

GHI = DHI + (DNI x COS(k-90))<br>
여기서 k 는 시간에 따른 각도 0 ~ 180 까지

```python
train[train.Hour == 12][['Hour', 'T']].describe().T
```

매일 12시의 기온을 통해 천정각에 가중치를 줄 수 있을까요..? 이건 나중에..

천정각을 계산하기가 어렵습니다. 계절별로 일출, 일몰시간도 다르고, 해가 떠 있는 시간도 다릅니다.<br>
일괄적으로 시간별 평균을 구해서, 모든 날을 똑같이 집어넣는 수 밖에 없을 것 같습니다.

```python
train.groupby('Hour')['DHI'].mean()
```

보통 빠르면 5시, 늦으면 6시에 해가 뜨고<br>
빠르면 16시, 늦으면 19시에 해가 집니다.<br>

저 데이터를 가지고 최소값 0, 최대값 180으로 정규화를 시키겠습니다.

```python
# DHI의 값에 따라서 0~180의 값을 가지는 정규화 함수입니다.

def angle_normalizor(list_, lags=48):
    min_ = min(list_)
    max_ = max(list_)
    return ((list_ - min_) / (max_ - min_))*180
```

```python
# 천정각에 의해 GHI를 출력하는 함수입니다.

def ghi_formula(DHI, DNI, theta):
    return (DHI + (DNI*np.cos((np.abs(theta-90)-1)*0.0174533)))
```

```python
# 하루 단위로(48개 index) DHI에 따른 천정각으로 GHI를 만들어주는 함수입니다.

def make_GHI(train, start_window=0, window=48):
    
    end_window = train.shape[0] / window
    #stride = window
    DHI = train.DHI # DHI value 값을 받습니다.
    DNI = train.DNI # DNI value 값을 받습니다.
    GHI = []
    
    #for i in tqdm(range(3)): # 테스트용 코드
    for i in range(np.int(end_window)):
        #print(i)
        if i == 0:
            #print(start_window)
            dhi_window = DHI[start_window:window] # 0부터 window=48 까지를 슬라이싱 해서 받습니다.
            dni_window = DNI[start_window:window]
            # 정규화 시키고,
            dhi_normal = angle_normalizor(dhi_window)
            # GHI 만들고
            new_GHI = ghi_formula(dhi_window, dni_window, dhi_normal)
            # 리스트에 더해줍니다.
            GHI = GHI + list(new_GHI)
        else:
            #print(window)
            dhi_window = DHI[(window*i):(window*(i+1))] # window*i 부터 window*(i+1) 까지를 슬라이싱
            dni_window = DNI[(window*i):(window*(i+1))]
            # 정규화 시키고,
            dhi_normal = angle_normalizor(dhi_window)
            # GHI 만들고
            new_GHI = ghi_formula(dhi_window, dni_window, dhi_normal)
            # 리스트에 더해줍니다.
            GHI = GHI + list(new_GHI)

            #window += stride # 윈도우를 옆으로 stride 시켜준다.
            
            
    train['GHI'] = GHI
    return train
```

### 1.3 새로운 GHI

기존에 만들었던 GHI가 DHI를 기반으로 만들어져서 노이즈가 조금 있습니다.<br>
이번 시간에 만들어 볼 변수는 5시부터 19시 까지 일괄적으로 넣어버리는 GHI,<br>
그리고 단순히 DHI+DNI= 총일사량 을 만들겠습니다.<br>
변수 중요도와 로스를 살펴보고 제출을 결정하겠습니다.

```python
train = train.drop('GHI', axis=1)
train
```

```python
# 시간에 따른 천정각 추가 함수
def time_sun_angle(train):
    
    one_radial = 6.17
    renew_radial = 0
    angle = []

    for i in range(48):
        if (i/2) > 19.1:
            angle.append(1)

        elif (i/2) > 4.9:
            renew_radial += one_radial
            angle.append(renew_radial)
        else:
            angle.append(1)

        i += 0.5
        
    train['sun_angle'] = angle*(int(train.shape[0]/48))
    
    return train
```

```python
# sun_angle 에 따른 GHI 만드는 함수
def sun_angle_ghi(train):
    train['new_ghi'] = train['DHI'] + (train['DNI']*(np.cos((train['sun_angle'] - 90)*0.0174533)))
    
    return train
```

```python
# 총 일사량 만들기
def total_irradiance(train):
    train['total_irradiance'] = train.DHI + train.DNI
    
    return train
```

트래인 테스트 셋 구성

```python
# 파생변수 만드는 함수
def make_features(train):
    print("original_shape: ", train.shape)
    # 가장 먼저 hour + minute 작업을 해줍니다.
    new_train = sum_hour_minute(train)
    # 필요 없는 day와 minute 을 지워줍니다.
    new_train.drop(['Day', 'Minute'], axis=1, inplace=True)  
    
    #-------------------- 파생변수 생성하면 이 아래에 추가
    # 일단 sun_angle을 만듭니다.
    new_train = time_sun_angle(new_train)
    # new_ghi 만듭니다.
    new_train = sun_angle_ghi(new_train)
    # total irradiance 만듭니다.
    new_train = total_irradiance(new_train)
    
    print("feature_creationed: ", new_train.shape)
    
    return new_train

# train set 만드는 코드
def make_train_dataset(train, shifted_columns, lags):
    print("original_shape: ", train.shape)
    
    # shift 칼럼을 만들어줍니다..
    new_train = shift_columns(train, shifted_columns, lags)
    print("shifted_columns: ", new_train.shape)
    
    # 타겟값을 마지막으로 만들어주겠습니다.
    new_train['TARGET_1'] = new_train.TARGET.shift(48)
    new_train['TARGET_2'] = new_train.TARGET_1.shift(48)
    
    # na 값을 떨궈줍니다.
    new_train.dropna(inplace=True)
    print("drop_na_columns: ", new_train.shape)

    X = new_train[:35000]
    val = new_train[35000:]
    print("new_train: ", new_train.shape)
    print("X : ", X.shape)
    print("val : ", val.shape)
    
    return X, val, new_train
    
def make_train_test_val(X, val, features):

    x_train = X[features]
    y_train_1 = X['TARGET_1']
    y_train_2 = X['TARGET_2']

    x_val = val[features]
    y_val_1 = val['TARGET_1']
    y_val_2 = val['TARGET_2']

    x_train.shape, y_train_1.shape, y_train_2.shape, x_val.shape, y_val_1.shape, y_val_2.shape
    
    return x_train, y_train_1, y_train_2, x_val, y_val_1, y_val_2
```
